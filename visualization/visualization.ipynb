```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
data = pd.read_csv('../data/comparison.csv')
```

```python
avg = data[['1', '2', '3', '4', '5']].mean(axis=1)
data.insert(5, "Average", avg, True)
```

```python
data
```

```python
this = data[data['Dataset'] == 'MIT']
colors = this['Data Type'].map({'Sequences': 'tab:blue', 'Thermodynamic Parameters': 'tab:red'})

plt.figure()
plt.bar(this.index, this['Average'], color=colors)
plt.title('Average r^2 by Data Type')
plt.ylabel('r^2')
plt.ylim(-1, 1)
plt.show()
```

```python
this = data[(data['Dataset'] == 'MIT') & (data['Data Type'] == 'Sequences')].groupby(['Model Type'])['Average'].mean()

plt.figure()
plt.bar(this.index, this.values)
plt.xticks(rotation=45)
plt.ylim(-0.25, 1)
plt.show()
```
