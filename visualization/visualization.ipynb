```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/comparison.csv')
```

```python
avg = data[['1', '2', '3', '4', '5']].mean(axis=1)
data.insert(5, "Average", avg, True)
```

```python
data
```

```python
this = data[(data['Dataset'] == 'MIT') & (data['Model Type'] != 'SGD')].reindex()
colors = this['Data Type'].map({'Sequences': 'tab:blue', 'Thermodynamic Parameters': 'tab:red'})

plt.figure()
plt.bar(range(12), this['Average'], color=colors)
plt.title('Average R^2 by Data Type (MIT)')
plt.xlabel('Model Trained')
plt.xticks(range(12), ['SVM', 'Dec. Tree', 'MLP', 'Rand. For.', 'Linear', 'Bayesian'] * 2, rotation=45)
plt.ylabel('R^2 value')
plt.show()
```

```python
this = data[(data['Dataset'] == 'Fernandez') & (data['Model Type'] != 'SGD')].reindex()
colors = this['Data Type'].map({'Sequences': 'tab:blue', 'Thermodynamic Parameters': 'tab:red'})

plt.figure()
plt.bar(range(12), this['Average'], color=colors)
plt.title('Average R^2 by Data Type (Fernandez)')
plt.ylabel('R^2 value')
plt.xlabel('Model Trained')
plt.xticks(range(12), ['SVM', 'Dec. Tree', 'MLP', 'Rand. For.', 'Linear', 'Bayesian'] * 2, rotation=45)
plt.show()
```
